## Antenna Beam

```python
def gain_dB(G_nu):
    return 10*log10(G_nu/amax(G_nu))

def mask_beam(gain_dB,min_dB,max_dB):
    return argwhere((gain_dB>min_dB) & (gain_dB<max_dB))
```

```python
g100_M=loadtxt('{}{}{}.txt'.format(path_M,'\Gains-', 100))
g100_B=loadtxt('{}{}{}.txt'.format(path_B,'\Gain_B_', 100))
g100_E=loadtxt('{}{}{}.txt'.format(path_E,'\Gain_E_', 100))
g100_In=loadtxt('{}{}{}.txt'.format(path_In,'\Gain_I_', 100))
g100_CA=loadtxt('{}{}{}.txt'.format(path_CA,'\Gain_CA_', 100))
g100_SA=loadtxt('{}{}{}.txt'.format(path_SA,'\Gain_SA_', 100))
g100_ISO=ones((91,360))
```

```python
gdB_Iso=gain_dB(g100_ISO)
gdB_M=gain_dB(g100_M)
gdB_B=gain_dB(g100_B)
gdB_E=gain_dB(g100_E)
gdB_In=gain_dB(g100_In)
gdB_CA=gain_dB(g100_CA)
gdB_SA=gain_dB(g100_SA)
```

```python
a,b=10.2,9.8
mask_10dB_Iso=mask_beam(gdB_Iso,-a,-b)
mask_10dB_M=mask_beam(gdB_M,-a,-b)
mask_10dB_B=mask_beam(gdB_B,-a,-b)
mask_10dB_E=mask_beam(gdB_E,-a,-b)
mask_10dB_In=mask_beam(gdB_In,-a,-b)
mask_10dB_CA=mask_beam(gdB_CA,-a,-b)
mask_10dB_SA=mask_beam(gdB_SA,-a,-b)
```

```python
i_ISO=argsort(radians(mask_10dB_Iso[:,1]))
i_M=argsort(radians(mask_10dB_M[:,1]))
i_B=argsort(radians(mask_10dB_B[:,1]))
i_E=argsort(radians(mask_10dB_E[:,1]))
i_In=argsort(radians(mask_10dB_In[:,1]))
i_CA=argsort(radians(mask_10dB_CA[:,1]))
i_SA=argsort(radians(mask_10dB_SA[:,1]))
```
